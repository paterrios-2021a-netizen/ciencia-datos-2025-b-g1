En este notebook se analizan las calificaciones de los estudiantes en diferentes materias y cortes acad√©micos.

**Caso:** La mejora total en el promedio de cada materia desde el Corte 1 hasta el Corte 3 y cu√°l de ellas present√≥ el mayor crecimiento en el rendimiento acad√©mico del grupo


```python
average_grades_corte1 = df[df['Corte'] == 1].groupby('Materia')['Nota Final'].mean()
average_grades_corte3 = df[df['Corte'] == 3].groupby('Materia')['Nota Final'].mean()

improvement_by_subject = average_grades_corte3 - average_grades_corte1

print("Mejora total en el promedio de cada materia (Corte 3 - Corte 1):")
display(improvement_by_subject)

best_subject = improvement_by_subject.idxmax()
max_improvement = improvement_by_subject.max()

print(f"\nLa materia con el mayor crecimiento en el rendimiento acad√©mico del grupo es: {best_subject} con una mejora de {max_improvement:.2f}")
```
